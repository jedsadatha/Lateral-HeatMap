## Data with Cluster Groups

```python
# other part
km = KMeans(n_clusters=6).fit(the_ml_data_old)
# create another dataframe
cluster_map = pd.DataFrame()

cluster_map['ID'] = df['subjectID']
cluster_map = cluster_map.set_index(the_ml_data_old.index)

cluster_map['cluster'] = km.labels_ ########## This is where we put in the clusters that were found
cluster_map['Gender'] = pd.Series(df['gender'])
cluster_map['Age'] = pd.Series(df['Age'])
cluster_map['FA'] = the_ml_data_old ## has FA values stored inide of it
cluster_map['Area'] = df['Order_FA_child']
cluster_map['AgeGroup'] = df['Age Group']
cluster_map['Category'] = df['category']
cluster_map = cluster_map.set_index(the_ml_data_old.index)
cluster_map['FA'] = the_ml_data_old ## has FA values stored inide of it

cluster_map
```

```python
# Creating Data for Cluster Data

C1 = cluster_map[cluster_map.cluster == 1]
C2 = cluster_map[cluster_map.cluster == 2]
C3 = cluster_map[cluster_map.cluster == 3]
C4 = cluster_map[cluster_map.cluster == 4]
C5 = cluster_map[cluster_map.cluster == 5]
C0 = cluster_map[cluster_map.cluster == 0]
```

```python
# Checking the numbers/amount of each clusters.

cluster_map['cluster'].value_counts()
```

# Heat and Cluster Maps (All Data)

## Catergory vs Age Group

```python
# Arranging Data
category_data_all = pd.pivot_table(cluster_map, values='FA',
                             index=['Category'],
                             columns='AgeGroup')

category_data_all.dropna(inplace=True)
category_data_all.index.names = ['Regions of the Brain (Category)']

# Clustermap Before Being Organized
category_all_clustermap = sns.clustermap(category_data_all);

# Rearranging the x-axis
groups = ['Child', 'Adolescent', 'Adult', 'Middle Age', 'Senior'] 

# reorder indexes/indices besed on the desired order
new_order = []
for group in groups:
    indexes = category_all_clustermap.data2d.columns.str.startswith(group)
    indexes_locs = np.where(indexes)[0].tolist()
    new_order += indexes_locs
    
# reorder df based on the new order
ordered_df_category_all = category_all_clustermap.data2d.iloc[:, new_order]
```

```python
# Clustermap After Being Organized

vcenter = 0
vmin, vmax = -0.02, 0.05 # max and min values from graph below
normalize = mcolors.TwoSlopeNorm(vcenter=vcenter, vmin=vmin, vmax=vmax)
colormap = sns.diverging_palette(20, 220, as_cmap=True)

## Run clustermap on the reordered dataframe by disabling 
## the clustering for both rows and columns

sns.clustermap(ordered_df_category_all,
              row_cluster=True,
              col_cluster=False,
              norm = normalize,
              method = 'ward',
              cmap=colormap).fig.suptitle('Laterality Index of All Data Clustered By Category and Age Group', 
                                       fontsize =15) 
# .fig.suptitle('Laterality Index of All Data Clustered By Area and Age Group', fontsize =15) 
```

```python
#Heat Map

sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(ordered_df_category_all, annot=True, cmap = colormap)
plt.title("Category vs Age Group (All Data)", fontsize =20)
```
